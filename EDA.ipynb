## 2. Data Gathering and Loading

```python
df = pd.read_csv('../data/clean_data')
df.head()
```

## 3.Uni-Variate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for the 'tip' variable
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='tip', color='Blue')
plt.title('Histogram Plot for tip')
plt.xlabel('tip')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for the 'tip' variable
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='tip', color='Blue')
plt.title('Violin Plot for tip')
plt.xlabel('tip')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

##  4. Bi-Variate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots for the univariate visualization
fig, axes = plt.subplots(1, 3, figsize=(22, 10))
fig.suptitle('Bivariate Visualization for tip vs. size', y=1.02, fontsize=20)

# Plotting the KDE plot
ax_kde = axes[0]
sns.kdeplot(data=df, x='size', y='tip', ax=ax_kde, color='r', fill=True, alpha=0.3)
ax_kde.set_title('KDE Plot for tip vs. size')
ax_kde.set(xlabel='Size', ylabel='tip')

# Plotting the scatter plot
ax_scatter = axes[1]
sns.scatterplot(data=df, x='size', y='tip', ax=ax_scatter, color='g', alpha=0.5)
ax_scatter.set_title('Scatter Plot for tip vs. size')
ax_scatter.set(xlabel='Size', ylabel='tip')

# Plotting the box plot
ax_box = axes[2]
sns.boxplot(data=df, x='size', y='tip', ax=ax_box, color='b')
ax_box.set_title('Box Plot for tip vs. size')
ax_box.set(xlabel='Size', ylabel='tip')

plt.tight_layout(pad=2.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot for 'tip' vs. 'day' with 'smoker' as hue
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='day', y='tip', hue='smoker', palette='Set1', s=100)
plt.title('Scatter Plot of tip vs. day (Colored by Smoker)')
plt.xlabel('Day of the Week')
plt.ylabel('tip')

plt.legend(title='Smoker', loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
